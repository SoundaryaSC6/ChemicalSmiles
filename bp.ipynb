```python
from rdkit import Chem
from rdkit.Chem import AllChem
from itertools import combinations
from rdkit import DataStructs
import networkx as nx
import matplotlib.pyplot as plt

smiles_list = ['COC(=O)C1CC2=CC(=O)CCC2(C)C23OC2CC2(C)C(CCC24CCC(=O)O4)C13',
               'CC(=O)SC1CC2=CC(=O)CCC2(C)C2CCC3(C)C(CCC34CCC(=O)O4)C12',
               'NCCCCC(NC(CCc1ccccc1)C(=O)O)C(=O)N1CCCC1C(=O)O',
               'CCOC(=O)C(CCc1ccccc1)NC(C)C(=O)N1CCCC1C(=O)O',
               'CCOC(=O)C(CCc1ccccc1)NC1CCc2ccccc2N(CC(=O)O)C1=O',
               'COc1cc2nc(N3CCN(C(=O)C4COc5ccccc5O4)CC3)nc(N)c2cc1OC',
               'COc1cc2nc(N3CCN(C(=O)c4ccco4)CC3)nc(N)c2cc1OC',
               'COc1cc2nc(N3CCN(C(=O)C4CCCO4)CC3)nc(N)c2cc1OC',
               'CCOC(=O)C1=C(COCCN)NC(C)=C(C(=O)OC)C1c1ccccc1Cl',
               'COC(=O)C1=C(C)NC(C)=C(C(=O)OC)C1c1ccccc1[N+](=O)[O-]',
               'NNc1nncc2ccccc12',
               'COCCc1ccc(OCC(O)CNC(C)C)cc1',
               'COc1ccc(CCN(C)CCCC(C#N)(c2ccc(OC)c(OC)c2)C(C)C)cc1OC',
               'CC(C)NCC(O)COc1ccc(CC(N)=O)cc1']

# Convert SMILES to RDKit molecules
mol_list = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]

# Calculate Morgan fingerprints for molecules
fps = [AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=1024) for mol in mol_list]

# Calculate similarity scores and store them in a dictionary
similarity_scores = {}
for i, j in combinations(range(len(fps)), 2):
    similarity = DataStructs.DiceSimilarity(fps[i], fps[j])
    similarity_scores[(i, j)] = similarity

# Set similarity threshold
threshold = 0.6

# Create a graph
G = nx.Graph()

# Add nodes (molecules) to the graph
for i in range(len(smiles_list)):
    G.add_node(i, label=smiles_list[i])  # Add node with molecule label

# Add edges above the threshold
for (i, j), similarity in similarity_scores.items():
    if similarity > threshold:
        G.add_edge(i, j, weight=similarity)  # Add edge with weight as similarity score

# Visualize the graph with clustering emphasized
pos = nx.spring_layout(G)  # Define the layout of the graph nodes

# Apply clustering algorithms (Girvan-Newman in this case)
communities_generator = nx.algorithms.community.girvan_newman(G)
top_level_communities = next(communities_generator)
colors = [0] * len(G.nodes())

# Assign different colors to different clusters for visualization
for i, comm in enumerate(top_level_communities):
    for node in comm:
        colors[node] = i

smiles_communities = []
for comm in top_level_communities:
    community_smiles = [smiles_list[node] for node in comm]
    smiles_communities.append(community_smiles)

# Print SMILES lists for each top-level community
for idx, community in enumerate(smiles_communities):
    print(f"Community {idx + 1} SMILES:")
    print(community)

# Draw the graph with nodes colored by clusters
nx.draw(G, pos, with_labels=True, node_size=500, cmap=plt.cm.rainbow, node_color=colors)

# Display the graph
plt.title("Molecule Graph with Clusters")
plt.show()
```

```python
import pandas as pd

# Load the dataset
# Replace 'file_path' with the actual path to your dataset file
file_path = r"C:\Users\sound\OneDrive\Desktop\Internshala work\New folder\20 drug dataset for combination.csv"
data = pd.read_csv(file_path)

drug_A_name = 'Prazosin'
drug_B_name = 'Terazosin'

# Filter the dataset for the two specific drugs
selected_drug_data = data[(data['Drug Name'] == drug_A_name) | (data['Drug Name'] == drug_B_name)]

# Print or further process the selected data for the two drugs
display(selected_drug_data)
```

```python
# Sort the data by drug name and probability to get top 5 targets for each drug
sorted_data_high = selected_drug_data.sort_values(by=['Drug Name', 'Probability*'], ascending=[True, False])
sorted_data_low = selected_drug_data.sort_values(by=['Drug Name', 'Probability*'], ascending=[True, True])

# Group by drug name and select the top 5 highest and lowest targets based on probability
top_5_highest = sorted_data_high.groupby('Drug Name').head(5)
top_5_lowest = sorted_data_low.groupby('Drug Name').head(5)

# Display or further process the top 5 highest and lowest targets for each drug
print("Top 5 Highest Probability Targets:")
display(top_5_highest)

print("\nTop 5 Lowest Probability Targets:")
display(top_5_lowest)
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Create a NetworkX graph
G = nx.Graph()

# Add root nodes (drugs)
drugs = ['Prazosin', 'Terazosin']
for drug in drugs:
    G.add_node(drug, node_type='drug')

# Function to add edges between drugs and their associated targets
def add_edges(df, ascending):
    for drug in drugs:
        targets = df[df['Drug Name'] == drug]['Target'].tolist()
        for target in targets:
            G.add_node(target, node_type='target')  # Assign node_type here
            G.add_edge(drug, target, edge_type='target', ascending=ascending)

# Add edges for top 5 highest and lowest targets
add_edges(top_5_highest, True)  # Highest probability targets
add_edges(top_5_lowest, False)  # Lowest probability targets

# Separate nodes by type for visualization
drug_nodes = [n for n, d in G.nodes(data=True) if d.get('node_type') == 'drug']  # Use .get() to handle KeyError
target_nodes = [n for n, d in G.nodes(data=True) if d.get('node_type') == 'target']  # Use .get() to handle KeyError

# Define positions for the nodes
pos = nx.spring_layout(G)

# Draw nodes and edges
plt.figure(figsize=(10, 8))
nx.draw_networkx_nodes(G, pos, nodelist=drug_nodes, node_color='r', node_size=500)
nx.draw_networkx_nodes(G, pos, nodelist=target_nodes, node_color='b', node_size=300)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
plt.title('Drug Targets Network Visualization')
plt.show()
```

```python
import pandas as pd

# Read the dataframes separately
pdock1 = pd.read_csv(r'C:\python projrct\python project 1\docking\Pdock1.csv', sep =' ')
pdock2 = pd.read_csv(r'C:\python projrct\python project 1\docking\Pdock2.csv', sep =' ')
pdock3 = pd.read_csv(r'C:\python projrct\python project 1\docking\Pdock3.csv', sep =' ')
tdock1 = pd.read_csv(r'C:\python projrct\python project 1\docking\dock 4.csv', sep =' ')
tdock2 = pd.read_csv(r'C:\python projrct\python project 1\docking\dock5.csv', sep =' ')
tdock3 = pd.read_csv(r'C:\python projrct\python project 1\docking\dock6.csv', sep =' ')

# Store each dataframe in a list
dataframes = [pdock1, pdock2, pdock3, tdock1, tdock2, tdock3]

# Store the minimum DeltaG for each dataframe in a list
min_deltaG_per_dataframe = []

# Calculate minimum DeltaG for each dataframe
for df in dataframes:
    min_deltaG = df.groupby(['ClusterRank'])['deltaG'].min().reset_index()
    min_deltaG_per_dataframe.append(min_deltaG)

# Display the minimum DeltaG for each dataframe
for idx, min_deltaG_df in enumerate(min_deltaG_per_dataframe, start=1):
    print(f"DataFrame {idx} - Minimum DeltaG:")
    print(min_deltaG_df)
    print()
```
